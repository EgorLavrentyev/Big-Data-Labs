```python
#task 1: sin cos plots
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#task 2, 3: x,y,z vectors
x = [i for i in range(-10, 6)]
y = [i for i in range(-5, 11)]
z = [(y[int(i / 2)] if i % 2 == 0 else x[int((i-1) / 2)]) for i in range(len(x) + len(y))]
print('vector z: ', z)
z.sort()
print('vector z, sorted: ', z)
```

```python
#task 4: calculate x,y,z norms 
from numpy.random import uniform
from math import sqrt

def first(vector:list):
    res = 0
    for el in vector:
        res += abs(el)
    return res

def second(vector:list):
    res = 0
    for el in vector:
        res += el**2
    return sqrt(res)      

def inf(vector:list):
    return max(vector) 

def weighted(vector:list):
    weights = [uniform(0, 1) for i in range(len(vector))]
    weights = [w /sum(weights) for w in weights]
    res = 0
    for i in range(len(vector)):
        res += abs(vector[i])*weights[i]
    return res

print('first x norm:', first(x))
print('second x norm:', second(x))
print('inf x norm:', inf(x))
print('weighted x norm:', weighted(x))
print()
print('first y norm:', first(y))
print('second y norm:', second(y))
print('inf y norm:', inf(y))
print('weighted y norm:', weighted(y))
print()
print('first z norm:', first(z))
print('second z norm:', second(z))
print('inf z norm:', inf(z))
print('weighted z norm:', weighted(z))
```

```python
#task 5: matrix operations
matrix = np.random.randint(1, 101, size=(10, 10))
row_sums = np.sum(matrix, axis=1)
column_sums = np.sum(matrix, axis=0)
print(matrix)
print("row sums:", row_sums, end='\n')

print("col_sums:", column_sums)

mul_row = matrix * row_sums[:, np.newaxis]
mul_col = matrix * column_sums[:, np.newaxis]

print ('matrix*row sum: \n', mul_row)
print ('matrix*col sum: \n', mul_col)

cut_matrix = matrix[:5, :5]

print (cut_matrix)
```

```python
#task 6: factorial
def factorial(n:int):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
print('input number to calculate factorial: ', factorial(int(input())))    
```
